```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx 

plt.close('all')
# gera um grafo aleat%*\color{codegreen}{ó}*)rio de Erd%*\color{codegreen}{ö}*)s-R%*\color{codegreen}{é}*)nyi com 1000 n%*\color{codegreen}{ó}*)s e probabilidade 0.01
ger = nx.gnp_random_graph(1000,0.01)
graus = dict(nx.degree(ger)).values() 

# calculo da centralidade de excentricidade
ec  = nx.eccentricity(ger)
cec = [1.0/ec[n] for n in ec]   

# calculo da centralidade de proximidade
# s%*\color{codegreen}{ó}*) salvamos os valores do dicion%*\color{codegreen}{á}*)rio e transformamos em lista
cc  = list(nx.closeness_centrality(ger).values())

# calculo da centralidade harm%*\color{codegreen}{ô}*)nica
# s%*\color{codegreen}{ó}*) salvamos os valores do dicion%*\color{codegreen}{á}*)rio e transformamos em lista
chu= list(nx.harmonic_centrality(ger).values())
ch = [x/(len(chu)-1) for x in chu]  #normaliza%*\color{codegreen}{çã}*)o
    
plt.figure()
plt.plot(graus,cec,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')

plt.figure()
plt.plot(graus,cc,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')

plt.figure()
plt.plot(graus,ch,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx 

# gera lista com valores de p
p = np.linspace(-1,201,1000)

# calculo do comprimento dos caminhos mais curtos desde n%*\color{codegreen}{ó}*) 1
sp= np.array(list(nx.shortest_path_length(ger,1).values()))

# calculo da centralidade generalizada para o n%*\color{codegreen}{ó}*) 1
cp= np.zeros_like(p)
for k in range(p.shape[0]):
    if p[k]:
        cp[k] = np.mean(sp[1:]**p[k])**(-1.0/p[k])
    else:
        cp[k] = np.prod(sp[1:])**(-1.0/(sp.shape[0]-1))
        
plt.figure()
plt.plot(p,cp,label='Centralidade generalizada')
plt.plot(p,np.ones_like(cp)*cec[1],label='Centralidade de excentricidade')
plt.plot(p,np.ones_like(cp)*cc[1], label='Centralidade de proximidade')
plt.plot(p,np.ones_like(cp)*ch[1], label='Centralidade harm%*\color{codepurple}{ô}*)nica')
plt.legend()
plt.xlabel('p')
plt.ylabel('centralidade')

plt.figure()
plt.plot(p,cp,label='Centralidade generalizada')
plt.plot(p,np.ones_like(cp)*cc[1], label='Centralidade de proximidade')
plt.plot(p,np.ones_like(cp)*ch[1], label='Centralidade harm%*\color{codepurple}{ô}*)nica')
plt.legend()
plt.xlabel('p')
plt.ylabel('centralidade')
plt.xlim((-1,3))
plt.ylim((0.31,0.36))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx 

plt.close('all')
# gera um grafo de Erdos-Rényi com 1000 nós e probabilidade 0.01
ger = nx.gnp_random_graph(1000,0.01)
graus = dict(nx.degree(ger)).values() 

# calculo da centralidade de autovetor
cev = list(nx.eigenvector_centrality_numpy(ger).values())

# calculo da centralidade de Katz
ck1 = list(nx.katz_centrality_numpy(ger,alpha=0.08,beta=1).values())
ck2 = list(nx.katz_centrality_numpy(ger,alpha=0.01,beta=1).values())
ck3 = list(nx.katz_centrality_numpy(ger,alpha=0.08,beta=2).values())

plt.figure()
plt.plot(graus,cev,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')
plt.ylim((-0.0025,0.0725))

plt.figure()
plt.plot(graus,ck1,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')
plt.ylim((-0.0025,0.0725))

plt.figure()
plt.plot(graus,ck2,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')
plt.ylim((-0.0025,0.0725))

plt.figure()
plt.plot(graus,ck3,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')
plt.ylim((-0.0025,0.0725))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx 

plt.close('all')
# gera um grafo aleat%*\color{codegreen}{ó}*)rio de Erd%*\color{codegreen}{ö}*)s-R%*\color{codegreen}{é}*)nyi com 1000 n%*\color{codegreen}{ó}*)s e probabilidade 0.01
ger = nx.gnp_random_graph(1000,0.01)
graus = dict(nx.degree(ger)).values() 

# calculo da centralidade de PageRank variando valor de \alpha
cpr1 = list(nx.pagerank(ger,alpha=0.85).values())
cpr2 = list(nx.pagerank(ger,alpha=0.65).values())
cpr3 = list(nx.pagerank(ger,alpha=0.25).values())

plt.figure()
plt.plot(graus,cpr1,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')
plt.ylim((0.0002,0.002))

plt.figure()
plt.plot(graus,cpr2,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')
plt.ylim((0.0002,0.002))

plt.figure()
plt.plot(graus,cpr3,'o')
plt.xlabel('grau')
plt.ylabel('centralidade')
plt.ylim((0.0002,0.002))

plt.show()
```
